```python
from sys import stdin
    
class Input:
    def __init__(self):
        self.lines = stdin.read().splitlines()
        self.count = 0
    
    def readline(self):
        for line in self.lines:
            yield line
    
    def line(self):
        try:
            return self.lines[self.count]
        finally:
            self.count += 1
    
    def array(self, sep = ' ', cast = int):
        return list(map(cast, self.line().split(sep = sep)))
    
    def known_tests(self):
        num_of_cases, = self.array()
    
        for case in range(num_of_cases):
            yield self
    
    def unknown_tests(self):
        while self.count < len(self.lines):
            yield self
    
class Problem:
    oo = float('inf')
    
    def __init__(self, input):
        n, r1, r2, r3, d = input.array()
        a = input.array()
        
        dp = [0, self.oo]
        for i in range(n):
            c0 = r1 * a[i] + r3
            c1 = min(r1 + r2, (a[i] + 2) * r1)
    
            if i + 1 == n:
                dp = [
                    min(dp[0] + c0, dp[0] + c1 + 2*d, dp[1] + c1, dp[1] + c0 - d),
                    self.oo
                ]
            else:
                dp = [
                    min(dp[0] + c0, dp[1] + min(c0, c1)),
                    dp[0] + c1 + 2*d
                ]
    
        print(d * (n - 1) + dp[0])
    
for tc in Input().unknown_tests():
    Problem(tc)
```
