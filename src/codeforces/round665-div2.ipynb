```python
from sys import stdin, stdout
    
class Input:
    def __init__(self):
        self.lines = stdin.read().splitlines()
        self.idx = 0
    
    def line(self):
        try:
            return self.lines[self.idx]
        finally:
            self.idx += 1

    def array(self, sep = ' ', cast = int):
        return list(map(cast, self.line().split(sep = sep)))

    def known_tests(self):
        num_of_cases, = self.array()
    
        for case in range(num_of_cases):
            yield self

    def unknown_tests(self):
        while self.idx < len(self.lines):
            yield self

'''
4 3 6 6 2 9
2 3 4 6 6 9
'''
class Problem:
    def __init__(self):
        self.oo = float("inf")
    
    def gcd(self, a, b):
        return a if b == 0 else self.gcd(b, a % b)

    def solve(self, n, a):
        b = sorted(a)
        c = []
        for i in range(len(a)):
            if a[i] != b[i]:
                c.append(b[i])
        ok = True
        for i in range(len(c)):
            if self.gcd(b[0], c[i]) != b[0]:
                ok = False
        print('YES' if ok else 'NO')

try:
    p = Problem()
    for tc in Input().known_tests():
        p.solve(*tc.array(), tc.array())
except Exception as e:
    import traceback
    traceback.print_exc(file=stdout) 
```
