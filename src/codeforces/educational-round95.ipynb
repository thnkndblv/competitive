```python
from sys import stdin, stdout
    
class Input:
    def __init__(self):
        self.lines = stdin.read().splitlines()
        self.idx = 0
    
    def line(self):
        try:
            return self.lines[self.idx]
        finally:
            self.idx += 1

    def array(self, sep = ' ', cast = int):
        return list(map(cast, self.line().split(sep = sep)))

    def known_tests(self):
        num_of_cases, = self.array()
    
        for case in range(num_of_cases):
            yield self

    def unknown_tests(self):
        while self.idx < len(self.lines):
            yield self

class Problem:
    def __init__(self):
        N = 200005
        self.oo = float("inf")
        self.dp = [[0, 0] for i in range(N)]
        self.dp[0][0] = self.oo

    def solve(self, n, a):
        for i in range(n):
            k1 = a[i]
            self.dp[i + 1][0] = self.dp[i][1] + k1
            self.dp[i + 1][1] = self.dp[i][0]

            if i > 0:
                k2 = a[i] + a[i - 1]
                self.dp[i + 1][0] = min(self.dp[i + 1][0], self.dp[i - 1][1] + k2)
                self.dp[i + 1][1] = min(self.dp[i + 1][1], self.dp[i - 1][0])
        
        print(min(*self.dp[n]))

try:
    p = Problem()
    for tc in Input().known_tests():
        p.solve(*tc.array(), tc.array())
except Exception as e:
    import traceback
    traceback.print_exc(file=stdout) 
```
